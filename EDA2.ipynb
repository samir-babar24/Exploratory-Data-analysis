```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
#NORMLIZATION
a=sns.load_dataset('titanic')
a
```

```python
a.isnull().sum()
```

```python
# here we can see that there are sum null values
#before going toward the normalization lets fill this stuff
anew=a
anew['age']=anew['age'].fillna(a['age'].mean)
anew['deck']=anew['deck'].fillna("D")
anew['embarked']=anew['embarked'].fillna("C")
anew['embark_town']=anew['embark_town'].fillna("Southampton")
anew.isnull().sum()
#now we have removed all null values
```

```python
#lets check which column is not normnlyy distributed and lets work on it
anew.describe()
```

```python
#we should work on sibsp,parch and fare column
#creating a new dataframe
df=anew[['sibsp','parch','fare']]
df.head(10)
```

```python
#here calling min max scaler
scaler=MinMaxScaler()
new_df=scaler.fit_transform(df.head(20))#fit transform convert values to 0 to1 values using normalization formula
new_df
```

```python
mydata=pd.read_csv("/content/house_price_regression_dataset.csv")
#mydata.head(200)
#detection of an outlier using function

def detection(mydata):
  detected=[]#creating dummy array to save  outlier values

  mydata_std=np.std(mydata)#std deviation
  mydata_mean=np.mean(mydata)#mean

  cutoff=mydata_std*3#cutoff to see ehichg is outlier


#cutoff ranges
  lower_range=mydata_mean-cutoff
  upper_range=mydata_mean+cutoff

  #to find outliers
  for i in mydata:
    if i>upper_range or i<lower_range:
      detected.append(i)
  return detected
```
