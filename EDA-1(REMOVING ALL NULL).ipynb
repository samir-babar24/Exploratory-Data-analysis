```python
#if we got some null values then we have to operate on this
#we can delete column by using reding.dropna(axis=1) this will drop columns
#reading.dropna(axis=0)this will dorp your row
```

```python
#now filling the null values taking a column having null vaalues
new=reading
new['Deaths/1M pop']=new['Deaths/1M pop'].fillna(reading['Deaths/1M pop'].mean())
new
```

```python
new.isnull().sum()#checking death\columns null values removed or not
```

```python
#types of filling null values
#forward filling=reading['columname].ffill
# baxkwardfilling=reading['columname].ffill lets try it
new1=new
new1['ActiveCases']=new1['ActiveCases'].ffill()#forward
new1['Serious,Critical']=new1['Serious,Critical'].bfill()#backward

#filling all remaining
new1['Tests/1M pop']=new1['Tests/1M pop'].fillna(new['Serious,Critical'].median())
new1['WHO Region']=new1['WHO Region'].fillna("Africa")
new1['Continent']=new1['Continent'].fillna("asia")
new1['Population']=new1['Population'].fillna("8.2222")
new1['NewCases']=new1['NewCases'].fillna("not any new cases")
new1['NewDeaths']=new1['NewDeaths'].fillna("not any new deaths")
new1['NewRecovered']=new1['NewRecovered'].fillna("not any new recovered")
new1['TotalCases']=new1['TotalCases'].fillna("not any total cases")
new1['TotalDeaths']=new1['TotalDeaths'].fillna("not any total deaths")
new1['TotalRecovered']=new1['TotalRecovered'].fillna("not any total recovered")
new1['TotalTests']=new1['TotalTests'].fillna(new['TotalTests'].mean())
new1['Tot Cases/1M pop']=new1['Tot Cases/1M pop'].fillna(new['Tot Cases/1M pop'].mean())



new1
```
