```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Load dataset
titanic = sns.load_dataset("titanic")
```

```python

# Handling Missing Values
titanic['age'].fillna(titanic.groupby('pclass')['age'].transform("median"), inplace=True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)



titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1
titanic['is_alone'] = (titanic['family_size'] == 1).astype(int)

```

```python
# Checking for missing values
print(titanic.isnull().sum())
```

```python


# Visualizations

plt.figure(figsize=(10, 5))
sns.heatmap(titanic.isnull(), cmap='coolwarm', cbar=False, yticklabels=False)
plt.title('Missing Data Heatmap')
plt.show()


```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='survived', data=titanic, palette='coolwarm')
plt.title('Survival Count')
plt.show()
```

```python

plt.figure(figsize=(8, 4))
sns.barplot(x='sex', y='survived', data=titanic, palette='coolwarm')
plt.title('Survival Rate by Gender')
plt.show()

```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='pclass', y='survived', data=titanic, palette='viridis')
plt.title('Survival Rate by Passenger Class')
plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.histplot(titanic['age'], bins=30, kde=True, color='blue')
plt.title('Age Distribution')
plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.histplot(titanic['fare'], bins=30, kde=True, color='green')
plt.title('Fare Distribution')
plt.show()

```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='embarked', y='survived', data=titanic, palette='magma')
plt.title('Survival Rate by Embarked Port')
plt.show()

```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlations')
plt.show()

```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='family_size', y='survived', data=titanic, palette='crest')
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(data=titanic, x="age", hue="survived", fill=True, common_norm=False, palette="coolwarm", alpha=0.5)
plt.title('Survival Distribution by Age')
plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x="pclass", y="fare", data=titanic, palette="coolwarm")
plt.title('Boxplot: Fare vs. Passenger Class')
plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x="survived", y="age", data=titanic, palette="muted", split=True)
plt.title('Age Distribution Among Survivors & Non-Survivors')
plt.show()


```

```python
print("\nKey Insights:")
print("Women and children had a higher survival rate.")
print("First-class passengers had better survival chances.")
print("Passengers traveling alone had lower survival probabilities.")
print("Higher ticket fares were associated with higher survival rates.")
print("People boarding at Cherbourg had better survival chances.")
```
